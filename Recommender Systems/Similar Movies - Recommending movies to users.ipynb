### Movie recommendation using MovieLens 100k dataset

#### Imports

```python
import pandas as pd
import numpy as np
```

#### Loading MovieLens project's data

```python
## Movie rating data
ratings_cols = ['user_id', 'movie_id', 'rating']
ratings_df = pd.read_csv('ml-100k/u.data', sep='\t', names=ratings_cols, usecols=(range(3)))

## User data
movies_cols = ['movie_id', 'title']
movies_df = pd.read_csv('ml-100k/u.item', sep="|", names=movies_cols, usecols=(range(2)), encoding='latin-1')

ratings_df = pd.merge(movies_df, ratings_df)
```

```python
ratings_df.head()
```

#### Creating a pivot table to cross data about all the users and all the ratings

```python
movie_ratings = ratings_df.pivot_table(index=['user_id'], columns=['title'],
                                      values='rating')
```

#### Particular movies that an user hasn't rated will receive a NaN value 

```python
movie_ratings.head()
```

#### Creating a correlation matrix to compute pairwise similarities between ratings

```python
corr_matrix = movie_ratings.corr()
corr_matrix.head()
```

#### Improving the correlation matrix by setting a minimun threshold to 100 samples per item. In practice, this will remove movies that were rated less than 100 times in order to get a more consistent result. Take into account that 100 is a totally arbitrary number in this example.

```python
corr_matrix = movie_ratings.corr(method='pearson', min_periods=100)
```

```python
corr_matrix.head()
```

#### Reading the ratings for a specific user, dropping movies that he hasn't rated

```python
ratings = movie_ratings.loc[0].dropna()
```

```python
ratings.head()
```

#### Improving the similarities by multiplying the similar movies' scores to the ratings that the user gave to other movies. 

For example: Movies that are similar to the ones the user gave a HIGH rating will have a high similarity, while similar movies to the movies that this user gave a LOW rating will have a low final score, refining our recommender system

```python
similarity_candidates = pd.Series()

for i in range(len(ratings.index)):
    print("Adding similarity for: " +ratings.index[i] + "...")
    
    sims = corr_matrix[ratings.index[i]].dropna()

    sims = sims.map(lambda x: x * ratings[i])
    
    similarity_candidates = similarity_candidates.append(sims)

print("Sorting...")

similarity_candidates.sort_values(inplace=True, ascending=False)
```

```python
similarity_candidates.head(10)
```

#### Grouping repeated movies together and adding up their similarities

```python
similarity_candidates=similarity_candidates.groupby(similarity_candidates.index).sum()
```

#### Sorting results

```python
similarity_candidates.sort_values(inplace=True, ascending=False)
```

```python
similarity_candidates.head(10)
```

#### filtering out movies that this user has already rated

```python
filtered_similarities = similarity_candidates.drop(ratings.index)
```

```python
filtered_similarities.head(10)
```

#### There we have it! This is a simple demonstration on how to use user data to recommend other stuff to him/her. In this project we're recommending movies, but the item itself can be anything you can imagine: videos, products, pages, etc.
